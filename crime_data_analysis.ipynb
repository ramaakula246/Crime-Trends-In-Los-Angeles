```python
# Clean column name first
df.columns = df.columns.str.strip().str.upper().str.replace(' ', '_')

# Convert DATE_OCC to datetime with correct format
df['DATE_OCC'] = pd.to_datetime(df['DATE_OCC'], format='%m/%d/%Y %I:%M:%S %p', errors='coerce')

# Extract Year
df['YEAR'] = df['DATE_OCC'].dt.year

```

```python
# Dropping duplicates
df = df.drop_duplicates()

# Standardizing column names
df.columns = df.columns.str.strip().str.upper().str.replace(' ', '_')

# Changing data types
df['DATE_OCC'] = pd.to_datetime(df['DATE_OCC'])

```

```python
# Checking missing values
print(df.isnull().sum())

# Dropping rows with missing values in important columns
df_clean = df.dropna(subset=['DATE_OCC', 'AREA_NAME'])
# Replacing missing values correctly
df['WEAPON_DESC'] = df['WEAPON_DESC'].fillna('UNKNOWN')
df['VICT_SEX'] = df['VICT_SEX'].fillna('UNKNOWN')
df['VICT_DESCENT'] = df['VICT_DESCENT'].fillna('UNKNOWN')
df['PREMIS_DESC'] = df['PREMIS_DESC'].fillna('UNKNOWN')
df['CROSS_STREET'] = df['CROSS_STREET'].fillna('NOT AVAILABLE')

```

```python
 #Data Visualization
#1. Principles of Visualization
#Use visualizations to show:
#Frequency of crimes
#Trends over time
#Crime locations
#2. Matplotlib Plots
# Crime count per year
df['YEAR'].value_counts().sort_index().plot(kind='bar', figsize=(10, 5), color='skyblue', edgecolor='black')
plt.title("Crime Count Per Year")
plt.xlabel("Year")
plt.ylabel("Number of Crimes")
plt.xticks(rotation=45)
plt.tight_layout()
plt.show()

```

```python
#3. Advanced Visualization with Seaborn
# Crime distribution by area
plt.figure(figsize=(12, 6))
sns.countplot(y='AREA_NAME', data=df, order=df['AREA_NAME'].value_counts().index)
plt.title('Crime Count by Area')
plt.show()

# Time trend of crimes
df['Month'] = pd.to_datetime(df['DATE_OCC']).dt.to_period('M')
monthly_crime = df['Month'].value_counts().sort_index()

monthly_crime.plot(kind='line', figsize=(14,6))
plt.title("Monthly Crime Trend")
plt.ylabel("Number of Crimes")
plt.show()
```

```python
#4. Customizing Visualizations
import seaborn as sns
import matplotlib.pyplot as plt

# Set seaborn style
sns.set_style("whitegrid")

# Create histogram for YEAR
plt.figure(figsize=(10, 6))
sns.histplot(df['YEAR'], bins=10, kde=True, color='skyblue')

# Labels and title
plt.title("Crime Frequency by Year")
plt.xlabel("Year")
plt.ylabel("Crime Count")
plt.tight_layout()
plt.show()
```

```python
#4. Weapon Usage Trends
import pandas as pd
import matplotlib.pyplot as plt
import seaborn as sns

# Get top 10 weapon types
top_weapons = df['WEAPON_DESC'].value_counts().head(10).reset_index()
top_weapons.columns = ['Weapon', 'Count']

# Plot
plt.figure(figsize=(12, 6))
sns.barplot(data=top_weapons, x='Weapon', y='Count', hue='Weapon', palette='magma', legend=False)

plt.title("Top 10 Weapons Used in Crimes")
plt.ylabel("Number of Incidents")
plt.xlabel("Weapon Type")
plt.xticks(rotation=45)
plt.tight_layout()
plt.show()

```
